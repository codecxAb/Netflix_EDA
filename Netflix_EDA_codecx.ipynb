##Analyze the Best Directors

```python
df['director']= df['director'].fillna('not specified')
df['director']
```

```python
director_list= pd.DataFrame()

director_list = df['director'].str.split(",", expand = True).stack()
director_list = director_list.to_frame()

director_list.columns= ['Directors']
directors = director_list.groupby('Directors').size().reset_index(name = 'Total shows')
directors = directors[directors.Directors!= 'not specified']
directors = directors.sort_values('Total shows')
directors
```

```python
top10directors = directors.tail(10)
top10directors
```

```python
director_barchart = px.bar(top10directors, x='Directors', y = 'Total shows', title= 'Top 10 Directors with Total shows Published')
director_barchart
```

##Analyze The top casts

```python
df['cast']= df['cast'].fillna('not specified')
cast_list= pd.DataFrame()

cast_list = df['cast'].str.split(",", expand = True).stack()
cast_list = cast_list.to_frame()

cast_list.columns= ['Casts']
casts = cast_list.groupby('Casts').size().reset_index(name = 'Total shows')
casts = casts[casts.Casts!= 'not specified']
casts = casts.sort_values('Total shows')
top10casts = casts.tail(10)

casts_barchart = px.bar(top10casts, x='Casts', y ='Total shows', title = 'Total shows Done by the Cast')
casts_barchart

```

##Analyze The Trend of Movies & Shows

```python
df_year = df[['type', 'release_year']]
df_year = df_year.rename(columns={ 'release_year': 'Release Year' , 'type': 'Type'})
df_year = df_year.groupby(['Release Year', 'Type']).size().reset_index(name='Total')
df_year = df_year[df_year['Release Year']>= 1950]

year_linechart = px.line(df_year, x= 'Release Year', y='Total', color='Type', title = 'Trend of Movies and Shows on Netflix')
year_linechart
```

##Analyze the movie Duration Time

```python
df['Movie Duration'] = df[df['type'].str.contains('Movie', case=False)]['duration'].str.replace(' min', '').astype(float)
df['TV Show Duration'] = df[df['type'].str.contains('TV Show', case=False)]['duration']
duration_mean = df['Movie Duration'].mean()
df['Movie Duration'].fillna(value= duration_mean, inplace =True)
df.head()
```

```python
df_duration = df[['release_year', 'Movie Duration']]
df_duration = df_duration.groupby('release_year')['Movie Duration'].sum().reset_index()
df_duration = df_duration[df_duration['release_year']>2014]
duration_line = px.line(df_duration, x='release_year', y ='Movie Duration')
print(df_duration)
duration_line.show()
```

```python
average_duration_by_year= df[['release_year', 'Movie Duration']]
average_duration_by_year = average_duration_by_year.groupby('release_year')['Movie Duration'].mean().reset_index()
duration_avg_line = px.line(average_duration_by_year, x='release_year', y ='Movie Duration')
print(average_duration_by_year)
duration_avg_line.show()

```

##Analyze the top movies & shows producing Countires

```python

df_country = df[['country', 'release_year']]
df_country= df_country[df_country['release_year']>2009]
df_country = df_country.groupby(['country','release_year']).size().reset_index(name = 'Total published')
df_country = df_country.groupby('country').size().reset_index(name='Total')
df_country = df_country.sort_values('Total')
df_countryTop20 = df_country.tail(20)
country_bar = px.bar(df_countryTop20, x = 'Total', y = 'country', title = 'Top 20 Countries of Netflix in Show Releasing')
country_bar.show()

```

##Analyze the Sentiment Of the Contents

```python
df_sentiment = df[['release_year', 'description']].copy()
df_sentiment['Sentiment']= ''
for index, row in df_sentiment.iterrows():
  description = row['description']
  testimonial = TextBlob(description)2
  polarity = testimonial.sentiment.polarity

  if polarity == 0:
    sentiment = 'Neutral'
  elif polarity <= 0:
    sentiment = "Negetive"
  else:
    sentiment = 'Positive'

  df_sentiment.loc[index, 'Sentiment']= sentiment
df_sentiment = df_sentiment[df_sentiment['release_year']>=2000]
df_sentiment = df_sentiment.groupby(['release_year', 'Sentiment']).size().reset_index(name= 'total')
sentiment_chart = px.bar(df_sentiment , x = 'release_year', y = 'total', color = 'Sentiment', title = 'Sentiment of the Content in Netflix')
sentiment_chart
```
